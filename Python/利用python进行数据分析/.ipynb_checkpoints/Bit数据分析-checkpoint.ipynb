# 利用python进行数据分析

## 引言

2011年，URL缩短服务bit.ly跟美国政府网站usa.gov合作，提供了一份从生成.gov或.mil短链接的拥挤那里收集的匿名数据。

时区统计(数据来自bit.ly的1.usa.gov)

```python
from pandas import DataFrame, Series
import numpy as np
import pandas as pd
from collections import Counter
from collections import defaultdict
import json
```

```python
path = "usagov_bitly_data2013-05-17-1368832207.txt"
records = [json.loads(line) for line in open(path)]
records[0]
```

```python
records[0]["tz"]
```

```python
# 不是所有记录都有时区字段
time_zones = [rec['tz'] for rec in records if 'tz' in rec]
time_zones[:10]
```

```python
# 利用函数对时区进行计数
def get_count(sequence):
    counts = defaultdict(int)  # 初始化所有值为0
    for x in sequence:
        counts[x] += 1
    return counts


counts = get_count(time_zones)
counts['America/New_York']
```

```python
len(time_zones)
```

```python
def top_counts(count_dict, n=10):
    """
    return a sorted list[n:] of dict
    """
    value_key_pairs = [(count, tz) for tz, count in count_dict.items()]
    value_key_pairs.sort()
    return value_key_pairs[-n:][::-1]


top_counts(counts)
```

```python
# 使用 collections.Counter 计数时区

counts = Counter(time_zones)
counts.most_common(10)
```

```python
# 使用pandas 时区计数
frame = DataFrame(records)
frame
```

```python
tz_counts = frame['tz'].value_counts()
tz_counts[:10]
```

```python
clean_tz = frame['tz'].fillna("Missing")
clean_tz[clean_tz == ''] = 'Unknown'
tz_counts = clean_tz.value_counts()
tz_counts[:10]
```

```python
tz_counts[:10].plot(kind='barh', rot=0)
```

```python
frame['a'][1]
```

```python
# 分离agent信息
results = Series([x.split()[0] for x in frame.a.dropna()])
results[:5]
```

```python
results.value_counts()[:8]
```

```python
# 区分windows和非windows
cframe = frame[frame.a.notnull()]
operating_system = np.where(cframe["a"].str.contains(
    "Windows"), "Windows", "Not Windows")
# 根据时区和操作系统列表对数据分组
by_tz_os = cframe.groupby(['tz', operating_system])
```

```python
# 通过size对分组结果进行计数，利用unstack对计数结果进行重塑
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts[:10]
```

```python
# 选取最常出现的时区，根据agg_counts中的行数构造一个间接索引数组
# 用于升序排列
indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

```python
# 通过take截取最后10行
count_subset = agg_counts.take(indexer)[-10:]
count_subset
```

```python
# 通过stacked=True，生成一张堆积条形图
count_subset.plot(kind='barh', stacked=True)
```

```python
# 各行规范化为“总计为1”
normed_subset = count_subset.div(count_subset.sum(1), axis=0)
normed_subset.plot(kind='barh', stacked=True)
```
